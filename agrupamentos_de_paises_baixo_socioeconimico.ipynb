# utilezeio a base de dados **basic1**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

```

```python
# Carregar a base de dados basic1
data= pd.read_csv('basic1.csv')
```

```python
data.info()
```

```python
#Garantindo que o DataSet será bidimensional
X = data[['x', 'y']].values
```

```python
#Criei uma função para calcular o método de Elbow
def plot_elbow_method(x):
    wcss = []
    max_clusters = 10
    for k in range(1, max_clusters):
        kmeans = KMeans(n_clusters=k, random_state=99)
        kmeans.fit(data)
        wcss.append(kmeans.inertia_)

    plt.figure(figsize=(8, 6))
    plt.plot(range(1, max_clusters), wcss, marker='o', linestyle='--')
    plt.title('Método do Cotovelo')
    plt.xlabel('Número de Clusters')
    plt.ylabel('WCSS')
    plt.show()
```

```python
#Plotei o resultado do cálculo do método Elbow para saber qual o melhor número de Clusters
plot_elbow_method(X)

```

```python
#Aplicando o K-Means com o número de clusters igual a 4
kmeans = KMeans(n_clusters=4, random_state=99)
y_kmeans = kmeans.fit_predict(X)
```

```python
#Visualizando os clusters(grupos)
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='magenta', marker='+')
plt.title('Agrupamento com K-Means')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
#Avaliando o agrupamento com a pontuação de silhueta
score = silhouette_score(X, y_kmeans)
print(f'Pontuação de Silhueta: {score:.2f}')
```

## Agrupamento de países por dados sócio-econômicos

https://www.kaggle.com/datasets/rohan0301/unsupervised-learning-on-country-data  LINK

Descrição do problema:
A HELP International conseguiu arrecadar cerca de US$ 10 milhões. Agora, o CEO da ONG precisa decidir como usar esse dinheiro de forma estratégica e eficaz. Então, o CEO precisa tomar a decisão de escolher os países que mais precisam de ajuda. Portanto, seu trabalho como cientista de dados é categorizar os países usando alguns fatores socioeconômicos e de saúde que determinam o desenvolvimento geral do país. Então, você precisa sugerir os países nos quais o CEO precisa se concentrar mais.

# **Objetivo:**
Categorizar os países usando fatores socioeconômicos e de saúde que determinam o desenvolvimento geral do país.

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
#DataSet sócio-econômicos dos países e o dicionário do DataSet
DataSet = pd.read_csv('Country-data.csv')
```

```python
DataSerDictionary = pd.read_csv('data-dictionary.csv')
```

```python
print(DataSet.head()) #Primeiras 5 entradas do DataSet
```

```python
DataSerDictionary #Verificando o dicionário do DataSet
```

# pre-processamento

```python
DataSet_clean = DataSet.dropna() #Removendo todas as tuplas com NaN
```

```python
#Selecionando apenas os atributos de nosso interrese
Y = DataSet_clean[['child_mort', 'exports', 'health', 'imports', 'income',
                   'inflation', 'life_expec', 'total_fer', 'gdpp']]
```

```python
#Aplicando uma normalização nos dados antes de trabalhamos com eles
scaler = StandardScaler()
y_scaled = scaler.fit_transform(Y)
```

# Metodo do cotovelo

```python
plot_elbow_method(Y)
```

# Utilizando kmeans com cluster 5

```python
#Aplicando o K-Means com o número de clusters igual a 5
kmeans = KMeans(n_clusters=5, random_state=99)
DataSet_clean['cluster'] = kmeans.fit_predict(y_scaled)
#Visualizando os clusters
plt.figure(figsize=(10, 6))
plt.scatter(DataSet_clean['gdpp'], DataSet_clean['child_mort'], c=DataSet_clean['cluster'], cmap='viridis', marker='o')
plt.xlabel('GDP per capita')
plt.ylabel('Mortalidade Infantil (por 1000 nascimentos)')
plt.title('Agrupamento de Países por Dados Socioeconômicos')
plt.colorbar(label='Cluster')
plt.grid(True)
plt.show()
```

```python
#Avaliando os clusters
silhouette_avg = silhouette_score(y_scaled, DataSet_clean['cluster'])
print(f'Pontuação média de silhueta: {silhouette_avg}')
```

```python
#Calculando as médias dos clusters
cluster_summary = DataSet_clean.groupby(['cluster','country'])[['child_mort','exports','health','imports','income','inflation','life_expec','total_fer','gdpp']].mean()
cluster_summary
```

```python
#Calculando pontos críticos que determinei
critico_mortalidade = cluster_summary['child_mort'].quantile(0.75)
critico_saude = cluster_summary['health'].quantile(0.75)
critico_inflation = cluster_summary['inflation'].quantile(0.75)
critico_life_expec = cluster_summary['life_expec'].quantile(0.75)
```

```python
#Pegando os clusters correspondes a cada ponto crítico
clusters_critico_mortalidade = cluster_summary[cluster_summary['child_mort'] > critico_mortalidade ]
clusters_critico_saude = cluster_summary[cluster_summary['health'] > critico_saude ]
clusters_critico_inflation = cluster_summary[cluster_summary['inflation'] > critico_inflation ]
clusters_critico_life_expec = cluster_summary[cluster_summary['life_expec'] < critico_life_expec ]
```

```python
#Resetando o index para desfazer o gurupo que eu tinha feito mais cedo
clusters_critico_mortalidade = clusters_critico_mortalidade.reset_index()
clusters_critico_saude = clusters_critico_saude.reset_index()
clusters_critico_inflation = clusters_critico_inflation.reset_index()
clusters_critico_life_expec = clusters_critico_life_expec.reset_index()
```

```python
#Recomendações com índice na alta taxa de mortalidade infantil
plt.figure(figsize=(16, 10))
plt.barh(clusters_critico_mortalidade['country'], clusters_critico_mortalidade['child_mort'], color='red')
plt.xlabel('Mortalidade Infantil (por 1000 nascidos vivos)')
plt.title('Países Recomendados com Alta Mortalidade Infantil')
plt.show()
```

```python
#Recomendações com índice no alto custo na saúde
plt.figure(figsize=(16, 10))
plt.barh(clusters_critico_saude['country'], clusters_critico_saude['health'], color='green')
plt.xlabel('Custo per capito na saúde')
plt.title('Países Recomendados com Alto Custo de Saúde')
plt.show()
```

```python
#Recomendações com índice na alta taxa de inflação
plt.figure(figsize=(16, 10))
plt.barh(clusters_critico_inflation['country'], clusters_critico_inflation['inflation'], color='darkviolet')
plt.xlabel('Taxa de Inflação')
plt.title('Países Recomendados com Alta Taxa de Inflação')
plt.show()
```

```python
#Diminuindo nossa amostra para apenas os 20 países com a menor expectativa de vida
clusters_critico_life_expec =  clusters_critico_life_expec.nsmallest(20, 'life_expec')
```

```python
#Recomendações com índice de baixa expectativa de vida
plt.figure(figsize=(16, 10))
plt.barh(clusters_critico_life_expec['country'], clusters_critico_life_expec['life_expec'], color='darkblue')
plt.xlabel('Expectativa de Vida')
plt.title('Países Recomendados com Baixa Expectativa de Vida')
plt.show()
```
